```python
from numpy import pi, arange, sin, linspace
from bokeh.models import LinearAxis, Range1d
from bokeh.plotting import output_file, figure, show
f1 = figure(x_range=rvolbond['Date'],title="Bonds and Vols", width=800,y_range=(-0.1,0.1))
f1.grid.grid_line_alpha=0.3
f1.xaxis.axis_label = 'Time'
f1.yaxis.axis_label = '% Change Price/Vols'
f1.extra_y_ranges = {"y2": Range1d(start = 0, end = 300)}
f1.add_layout(LinearAxis(y_range_name = "y2"), 'right')


f1.line(rvolbond.index, rvolbond['TLT'].pct_change(), color='blue', legend = 'TLT')
f1.line(rvolbond.index, rvolbond['EM Bond'].pct_change(), color='green', legend = 'EM Bond')
f1.line(rvolbond.index, rvolbond['ICE Move BofA'], color='red', legend = 'Vols', y_range_name="y2")

f1.legend.location='top_left'

show(f1)
```
